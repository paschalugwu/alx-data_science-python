# Write a function to calculate the temperature range (Max_temperature_C - Min_temperature_C) for each farmer's field. Then, call the function with the following `Field_ID`: `1458`, `1895`, and `5443`. What are the results of these 3 calls?

````python
def calculate_temperature_range(field_id):
    # Filter rows for the given Field_ID
    field_data = df[df['Field_ID'] == field_id]

    # Calculate temperature range
    temp_range = field_data['Max_temperature_C'] - field_data['Min_temperature_C']

    return temp_range

# Call the function for the specified Field_IDs
field_1458_range = calculate_temperature_range(1458)
field_1895_range = calculate_temperature_range(1895)
field_5443_range = calculate_temperature_range(5443)

print(f"Temperature range for Field 1458:")
print(field_1458_range)
print(f"Temperature range for Field 1895:")
print(field_1895_range)
print(f"Temperature range for Field 5443:")
print(field_5443_range)
````

# What does the following code achieve?

- A. Sorts crops_types from the average lowest to highest minimum temperatures.
- B. Identifies the crop type with the lowest average minimum temperature.
- C. Identifies the crop type with the highest average minimum temperature
- D. Identifies the minimum temperature for each crop type.

`The given code snippet identifies the crop type with the lowest average minimum temperature. It iterates through each unique crop type in the dataset, calculates the mean minimum temperature for that crop, and keeps track of the crop type with the lowest average minimum temperature. The final result stored in variable c represents the crop type with the lowest average minimum temperature. Therefore, the correct answer is B.`

````python
a = df['Crop_type'].unique()

b = float('inf')

c = ''

for crop in a:

    d = df[df['Crop_type'] == crop]['Min_temperature_C'].mean()

    if d < b:

        b = d

        c = crop

print(c)
````

# Using Numpy, calculate the standard deviation of the 'Rainfall' for plots where the 'Plot_size' is larger than the median plot size of the dataset (round to 2 decimal places).

````python
# Import NumPy
import numpy as np

# Calculate the median plot size
median_plot_size = np.median(df['Plot_size'])

# Filter rows where Plot_size is larger than the median
filtered_df = df[df['Plot_size'] > median_plot_size]

# Calculate the standard deviation of Rainfall for filtered rows
rainfall_std = np.std(filtered_df['Rainfall'], ddof=0)  # ddof=0 for population standard deviation

# Round the result to 2 decimal places
rainfall_std_rounded = round(rainfall_std, 2)

print(f"The standard deviation of Rainfall for plots with Plot_size larger than the median is {rainfall_std_rounded}.")
````

# If you concatenate the first three digits of the most common â€˜Max_temperature_Câ€™ with the last three letters of the least common 'Crop_type', what string do you get?
`Note: Use the first mode if there are multiple modes`

````python
# Find the most common Max_temperature_C
most_common_temp = df['Max_temperature_C'].mode().iloc[0]

# Find the least common Crop_type
least_common_crop = df['Crop_type'].value_counts().idxmin()

# Extract the first three digits from most_common_temp
first_three_digits = str(int(most_common_temp))[:3]

# Extract the last three letters from least_common_crop
last_three_letters = least_common_crop[-3:]

# Concatenate the parts
result_string = first_three_digits + last_three_letters

print(f"The resulting string is: {result_string}")
````

# Assuming each 'Crop_type' contributes an integer value equal to its length (e.g., 'wheat' contributes 5), write a recursive function to sum the integer values for each unique crop type in the dataset. What is the sum?

````python
import pandas as pd

# Assuming you have already loaded your data into a DataFrame called 'df'
# You can replace 'df' with your actual DataFrame name

# Get unique crop types
unique_crop_types = df['Crop_type'].unique()

print(f"Unique crop types: {unique_crop_types}")
````

````python
def sum_crop_lengths(crop_types):
    if not crop_types:
        return 0
    else:
        # Get the length of the first crop type and add it to the sum
        return len(crop_types[0]) + sum_crop_lengths(crop_types[1:])

# Example data (replace with your actual data)
crop_types = ['coffee' 'potato' 'banana' 'wheat' 'cassava' 'maize' 'tea' 'rice']

# Calculate the sum of integer values
total_sum = sum_crop_lengths(crop_types)

print(f"The sum of integer values based on crop lengths is: {total_sum}")
````

# Write Python code to perform a t-test comparing the average 'Annual_yield' between 'coffee' and 'banana' crop types using scipy.stats. What is the p-value, rounded to three decimal places?

````python
import pandas as pd
from scipy.stats import ttest_ind

# Assuming you have already loaded your data into a DataFrame called 'df'
# You can replace 'df' with your actual DataFrame name

# Filter data for 'coffee' and 'banana' crop types
coffee_data = df[df['Crop_type'] == 'coffee']['Annual_yield']
banana_data = df[df['Crop_type'] == 'banana']['Annual_yield']

# Perform t-test
t_statistic, p_value = ttest_ind(coffee_data, banana_data, equal_var=False)

# Round p-value to three decimal places
rounded_p_value = round(p_value, 3)

print(f"P-value: {rounded_p_value}")
````
